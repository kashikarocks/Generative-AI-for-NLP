**Basic feedforward neural network for binary classification.**

```python
import numpy as np #especially used for working with arrays and matrices.
np.random.seed(0) #"Start your random number generation from this fixed point."
X = np.random.rand(4, 2) #This function generates a 4×2 matrix (array) or 4rows & 2columns
X 
```

```python
import numpy as np #especially used for working with arrays and matrices.
from keras.models import Sequential #Sequential means the model is a linear stack of layers, where each layer
                                    #has exactly one input and one output.
from keras.layers import Dense #Each neuron in this layer receives input from every neuron in the previous layer.
```

```python
# Step 2: Generate some random data for demonstration
np.random.seed(0)
X = np.random.rand(100, 2)  # Features (2 input neurons)
```

```python
y = (X[:, 0] + X[:, 1] > 1).astype(int)  # Binary target variable 
# Adds the 2 features in each row: X[:, 0] + X[:, 1]
#If the sum is greater than 1, the result is True (→ 1)else 0
#.astype(int) converts the boolean array to integers
```

```python
y
```

# Create a Sequential model

```python
## Step 3: Create a Sequential model
first_model = Sequential()
```

A Dense layer is a fully connected layer (every neuron is connected to all outputs of the previous layer). This one has 3 neurons.

input_dim=2	The input layer has 2 features (or 2 input neurons). This is required only for the first layer of the model, so it knows what shape to expect.

```python
# Step 4: Add layers to the model
first_model.add(Dense(3, input_dim=2, activation='relu'))  # Hidden layer with 3 neurons
first_model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron (binary classification)
```

```python
# Step 5: Compile the model
first_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#Prepares the model to be trained using .fit()
#loss='binary_crossentropy'	Loss function for binary classification.
#Optimizer that updates model weights to minimize loss.
```

**Epoch** refers to one complete pass of the entire training dataset through the learning algorithm. Essentially, it's how many times the algorithm sees and learns from the complete set of training data. 

```python
# Step 6: Train the model
first_model.fit(X, y, epochs=12, batch_size=20)
```

**So, your training dataset has about 100 samples** (since 5 batches × 20 = 100).

  - 5th batch: samples 80–99

```python
# Step 7: Evaluate the model
loss, accuracy = first_model.evaluate(X, y)
print(f'Loss: {loss:.4f}')
print(f'Accuracy: {accuracy*100:.2f}%')
```

**Decreasing batch_size:**

**Pros:**

1) More frequent updates = potentially better convergence

2) Introduces noise in learning → helps generalize

**Cons:**

1) Training takes longer

2) Less stable learning (loss may fluctuate more)

**Increasing batch_size:**

**Pros:**

1) Faster training (parallelism, fewer weight updates)
2) More stable updates (averaged over more data)

**Cons:**

1) May miss subtle patterns (less noise = less learning flexibility)

2) May require more memory

3) Can result in lower generalization performance

# Activation Fxns

**Types of activation fxns:**
1) Sigmoid fxn 

2) Rectified Linear Unit(ReLU)

3) Leaky ReLU

4) Softmax(for multi-class classification)

# 1) Sigmoid fxn

Range: (0, 1) 

Pros: Smooth, interpretable as probabilities, historically popular.

Cons: Suffers from vanishing gradients problem (gradients become very small), not used much in hidden layers anymore.

# 2) Rectified Linear Unit(ReLU)

Range: [0, +∞)

Pros: Fast to compute, helps with the vanishing gradients problem, widely used in hidden layers

Cons: Can suffer from the dying ReLU problem (neurons can get stuck during training and stop learning).

Formula: F(x) = max(0,x)

**Note:-** ReLU's simplicity and efficiency have made it a default choice in many deep learning models, especially convolutional neural networks (CNNs). 

# 3) Leaky ReLU

Range: (-∞, +∞)

Pros: Addresses the dying ReLU problem by allowing a small gradient for negative inputs.

Cons: Slightly more computationally expensive than ReLU.

Formula: f(x)=max(0.01x,x) (or a small, non-zero slope for negative values

# 4) Softmax (for multi-class classification)

Range: (0, 1), and the sum of all outputs is 1.

Purpose: Scales the outputs so that they can be interpreted as probabilities, commonly used in the output layer 
for multi-class classification.

Generally, ReLU or its variants are preferred in hidden layers due to their computational efficiency and better performance in practice.

Sigmoid or softmax is commonly used in the output layer for binary or multi-class classification task

# Loss fxns

Loss functions quantify the error or discrepancy between the predicted output of the neural network and the actual target values during training. 

**Types**

1) Mean Squared Error (MSE):- Suitable for regression tasks.

2) Binary Cross-Entropy Loss (Log loss) :- Suitable for binary classification tasks.

3) Categorical cross-Entropy:- Suitable for multi-class classification tasks.

In Keras, as well as in other machine learning frameworks, **AUC stands for Area Under the ROC Curve**. It's a metric **used to evaluate the performance of a binary classification model**. The ROC (Receiver Operating Characteristic) curve is a graphical representation of the trade-off between the true positive rate (sensitivity) and the false positive rate (1 - specificity) for different threshold values. The AUC value represents the area under this curve, which ranges from 0 to 1.

```python
import tensorflow as tf

# Assuming y_true and y_pred are your true labels and predicted probabilities respectively
y_true = [0, 1, 1, 0]
y_pred = [0.9, 0.9, 0.8, 0.2]

auc = tf.keras.metrics.AUC()
auc.update_state(y_true, y_pred)
print("AUC Score:", auc.result().numpy())
```
