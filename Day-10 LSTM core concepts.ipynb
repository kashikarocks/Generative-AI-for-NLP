# Understanding LSTM(Long Short-Term Memory)  Networks

1) Long Short Term Memory networks – usually just called **“LSTMs”** – are a **special kind of RNN**, capable of learning long-term dependencies or are explicitly designed to avoid the long-term dependency problem. 

2) Remembering information for long periods of time is practically their default behavior, not something they struggle to learn!

**Standard RNN Architecture**:- 
1) All recurrent neural networks have the form of a chain of repeating modules of neural network. 

2) In standard RNNs, this repeating module will have a very simple structure, such as a single tanh layer.

**LSTM Architecture**

LSTMs also have this chain like structure, but the repeating module has a different structure. Instead of having a single neural network layer, there are four, interacting in a very special way.

1) It consists of a memory cell that can maintain information over long sequences, controlled by three gates: forget gate, input gate, and output gate.

2)  The forget gate decides what information to discard from the cell state.
    
3)  The input gate decides what new information to store in the cell state.

4)  The output gate decides what information to output from the cell state.

5)  LSTM's ability to retain and forget information over long periods makes it effective for tasks involving sequential data with long-term dependencies.
