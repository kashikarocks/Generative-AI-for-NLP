# Tokenization in NLP

Tokenization is a fundamental step in Natural Language Processing (NLP). It involves dividing a 
Textual input into smaller units known as tokens. These tokens can be in the form of words,
characters, sub-words, or sentences. It helps in improving interpretability of text by
different models. 

# What is Tokenization in Natural Language Processing (NLP)?

NLP is a subfield of Artificial Intelligence,information engineering,and human-computer interaction. 
It focuses on how to process and analyze large amounts of natural language data efficiently.
It is difficult to perform as the process of reading and understanding languages is far more 
complex than it seems at first glance.

   1) Tokenization is a foundation step in NLP pipeline that shapes the entire workflow.
   
   2) Involves dividing a string or text into a list of smaller units known as tokens.
   
   3) Uses a tokenizer to segment unstructured data and natural language text into distinct chunks of information, treating them as different elements.
   
   4) Tokens: Words or Sub-words in the context of natural language processing. 
      **Example:** A word is a token in a sentence, A character is a token in a word, etc.
   
   5) **Application:** Multiple NLP tasks, text processing, language modelling, and machine translation.

# Part of Speech(POS) Tagging

**Part-of-speech (POS) tagging** is a fundamental process in Natural Language Processing (NLP) where each word in a text is assigned a **grammatical category**, such as noun, verb, adjective, etc.This process helps computers understand the syntactic structure and meaning of sentences.

```python
import nltk
from nltk.tokenize import PunktSentenceTokenizer
```

```python
s = 'The sky is blue \n The cat chased the mouse.'
```

```python
stn = nltk.sent_tokenize(s)   
for sent in stn: 
    print(nltk.pos_tag(nltk.word_tokenize(sent)))
```
