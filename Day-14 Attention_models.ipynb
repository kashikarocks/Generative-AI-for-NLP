# Attention Models

An attention model assigns varying levels of importance, or "attention scores," to different elements (e.g., words, phrases) within an input sequence based on their relevance to the current task or the generation of a specific output element. This allows the model to create a "context vector" that is a weighted sum of the input elements, where the weights are determined by the attention scores.

In sequence-to-sequence models for machine translation, attention allows the model to focus on the most relevant words in the source language sentence when generating each word in the target language. For instance, when translating "The cat sat on the mat" to French, the model might pay more attention to "cat" when generating "chat" and to "mat" when generating "tapis."
